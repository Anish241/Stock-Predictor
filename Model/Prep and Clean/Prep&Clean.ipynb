```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Dataset/GR2.csv')
df.head()
```

```python
time = []
for i in range(len(df)):
    time.append(i)
df['Time'] = time
# Remove the time from the date column
df['Date'] = df['Date'].apply(lambda x: x.split(' ')[0])
# Convert the date column to datetime format input format is DD/MM/YYYY
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df.head()
```

```python
prev_close = df['Close'].shift(1)
df['Prev Close'] = prev_close
df.head()
```

```python
prev_open = df['Open'].shift(1)
df['Prev Open'] = prev_open
df.head()
```

```python
prev_opdiff = df['Prev Open'] - df['Prev Close']
df['PrevOPDiff'] = prev_opdiff
df.head()
```

```python
df["Prev_High"] = df["High"].shift(1)
df["Prev_Low"] = df["Low"].shift(1)
df["Prev_Volume"] = df["Volume"].shift(1)
df["Prev_Open"] = df["Open"].shift(1)
df["Prev_Close"] = df["Close"].shift(1)
df["CloseChange"] = df["Close"] - df["Prev_Close"]
df["HighChange"] = df["High"] - df["Prev_High"]
df["Prev_CO_Diff"] = df["Prev_Close"] - df["Prev_Open"]
df["Prev_HL_Diff"] = df["Prev_High"] - df["Prev_Low"]
df["isUp"] = df["Close"] > df["Prev_Close"]
# drop first row
df = df.drop(df.index[0])
df = df.reset_index()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df.to_csv('../Dataset/GR2CLEAN.csv', index=False)
```
